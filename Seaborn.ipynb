```python
# for both numerical attributes
sns.relplot(x="Age", y="Fare", data = df)
```

```python
# for both numerical and catagorical
sns.catplot(x="Embarked",y="Age",data=df)
```

```python
sns.catplot(x="Embarked",y="Age",jitter = False, hue="Survived", data=df)
```

```python
  sns.catplot(x="Embarked",y="Age", kind = "swarm", data=df)
```
