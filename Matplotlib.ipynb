```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Draw a line in a diagram from position (1,5) to position (2,10)
xpoints = np.array([1,2])
ypoints = np.array([5,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints = np.array([1,20])
ypoints = np.array([10,200])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints = np.array([1,6])
ypoints = np.array([4,5])
plt.plot(xpoints,ypoints,'*')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])

plt.plot(ypoints,'*''r')
plt.show()
```

```python
# ms--> marker size
ypoints = np.array([3,8,1,12])
xpoints = np.array([5,6,7,9])
plt.plot(xpoints,ypoints,marker = '*',ms = 30,color='red')
plt.show()
```

```python
# mec --> marker edge color
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker = "o", ms = 30 , mec = 'k',color = 'pink')
plt.show()
```

```python
# mfc --> marker face color
ypoints = np.array([3,8,1,10])

plt.plot(ypoints,marker = 'o', ms = 20, mfc='g',mec='k')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, linestyle = 'dashed', marker='o',mec = 'hotpink', ms=20 , mfc = 'blue', linewidth=3)
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dotted', marker = 'o', ms = 25,
         mfc = "green" , mec = "red")
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y,color="r",linewidth=1)

plt.xlabel("Average Pulse",fontsize=20,color="red")
plt.ylabel("Calorie Burnage",fontsize=10, color="g")
plt.title("Sample",fontsize=30,color="b")

plt.show()
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y,color="orange",linewidth=1)

plt.xlabel("Average Pulse",fontsize=10,color="red")
plt.ylabel("Calorie Burnage",fontsize=10, color="g")
plt.title("Sample",fontsize=30,color="b")

plt.grid(color="green", linestyle="dotted") # for only one axis axis='y' or axis='x'

plt.show()
```

```python
y = np.array([10,25,15,19])
mylabels = ["Cherries","Apples","Bananas","Dates"]

plt.pie(y, labels = mylabels, startangle = 180)
plt.show()
```

```python
y = np.array([10,25,15,19])
mylabels = ["Cherries","Apples","Bananas","Dates"]
myexplode = [0,0.9,0.75,0.2]
mycolors = ["red","green","black","yellow"]

plt.pie(y, explode = myexplode, colors = mycolors)
plt.legend(labels = mylabels)
plt.show()
```
